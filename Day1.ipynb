# Day 1, part 1

```python
# Read lines without newline characters
with open('Day1.txt', 'r') as file:
    lineList = [line.strip() for line in file]

print(lineList)
```

```python
item1List = []
item2List = []

for item in lineList:
    item1 = int(item.split(" ")[0])
    item2 = int(item.split(" ")[3])
   
    item1List.append(item1)
    item2List.append(item2)

item1List.sort()
item2List.sort()

print(item1List)
print(item2List)
```

```python
total = 0

for num in range(0, len(item1List)):
    diff = abs(item1List[num] - item2List[num])
    total += diff

print(total)
```
